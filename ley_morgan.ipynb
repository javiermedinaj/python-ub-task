```python

import matplotlib.pyplot as plt

A = {2, 4, 6, 20}
B = {1, 7, 13, 20}
C = {0, 5, 20}


interseccion_ABC = A.intersection(B).intersection(C)

# Grafico
plt.figure(figsize=(6, 6))
venn = plt.gca()
venn.set_title('Intersecci√≥n de A, B y C')
venn.add_patch(plt.Circle((0.3, 0.5), 0.3, color='blue', alpha=0.5, label='A'))
venn.add_patch(plt.Circle((0.6, 0.5), 0.3, color='red', alpha=0.5, label='B'))
venn.add_patch(plt.Circle((0.45, 0.2), 0.3, color='green', alpha=0.5, label='C'))
venn.text(0.45, 0.5, interseccion_ABC, color='black', fontsize=12, ha='center', va='center')
plt.axis('off')
plt.legend(loc='upper right')
plt.show()
```
